CREDIT CARD FRAUD DETECTION

Importing Libraries 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Read a CSV file named "creditcard.csv" into a pandas DataFrame and store it in the 'data' variable.
data = pd.read_csv("creditcard.csv")
data
```

```python
data.head(9)
# Display the first 9 rows of the DataFrame 'data' to examine its contents.
```

```python
data.describe() # Generate summary statistics for the DataFrame 'data' to get an overview of its numerical data.
```

```python
data.isnull() # Check for missing values in the DataFrame 'data' and create a new DataFrame of boolean values.
```

```python
data.isnull().sum() # Calculate the number of missing values in each column of the DataFrame 'data'.
```

```python
data['Class'].value_counts()
```

```python
norm = data[data.Class==0]
norm.shape
```

```python
fraud = data[data.Class==1]
fraud.shape
```

```python
norm.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby('Class').mean()
```

```python
norm_sample = norm.sample(n=492)
```

```python
new_data = pd.concat([norm_sample,fraud],axis=0)
```

```python
new_data.head(10)
```

```python
new_data.groupby('Class').mean()
```

```python
X = new_data.drop(columns='Class',axis=1)
```

```python
Y=new_data['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_pred = model.predict(X_train)
```

```python
train_data_accuracy = accuracy_score(X_train_pred,Y_train)*100
```

```python
train_data_accuracy
```

```python
X_test_pred = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_pred,Y_test)*100
```

```python
test_data_accuracy
```
